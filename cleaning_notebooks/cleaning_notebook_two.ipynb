creating a function to classify anxiety into (Normal, Mild, Moderate, Severe)

creating a function to interpret the HADS depression total, though the HADS is a screening tool not a diagnostic tool the total of each subscale could be classified into 3 main classes:
1. 0-7: normal
2. 8-10 : possible anxiety/ depression
3. 11 and above: anxiety/depression is likely present 

namings that are used here:

No depression

possible depression

probable clinical depression (the worst case)

```python
def depression_diagnosis_type(score):
    if score <= 7:
        return "No Depression"  # 0
    elif score >= 8 and score <= 10:  # 1
        return "Possible Depression"
    else:  # 2
        return "Probable Clinical Depression"
```

```python
df["depression_diagnosis_type"] = df["HADS_DEP_Scale_Total"].apply(
    depression_diagnosis_type
)
```

```python
df
```

creating the same function described above for the HADS anxiety subscale

```python
def anxiety_diagnosis_type(score):
    if score <= 7:
        return "No anxiety"  # 0
    elif score >= 8 and score <= 10:  # 1
        return "Possible anxiety"
    else:  # 2
        return "Probable Clinical anxiety"
```

```python
df["anxiety_diagnosis_type"] = df["HADS_ANX_Scale_Total"].apply(anxiety_diagnosis_type)
```
